```python
dfreg = df
dfreg['sales_method'] = dfreg['sales_method'].replace('Email', 2).replace('Call', 1).replace('Email & Call', 3)
dfreg.head()
```

```python
dfreg['week2'] = dfreg['week'] **2
dfreg['years_as_customer2'] = dfreg['years_as_customer'] ** 2
dfreg.head()
```

```python
formular = 'revenue ~ week + week2 + C(sales_method)'
model = smf.ols(formular, data = dfreg).fit()
model.params
```

```python
dfp = pd.DataFrame()
dfp['week'] = np.linspace(1, 6, num = 15000)
dfp['week2'] = dfp['week'] ** 2
dfp['sales_method'] = 3
dfp['sales_method2'] = dfp['sales_method'] ** 2
dfp = dfp.round()
dfp.head()
```

```python
dfp = pd.DataFrame()
dfp['week'] = np.linspace(1, 6, num = 4962)
dfp['week2'] = dfp['week'] ** 2
dfp['sales_method'] = 1
dfp['sales_method2'] = dfp['sales_method'] ** 2
pred = model.predict(dfp)
dfp['pred'] = pred
call_g = dfp.groupby('week')['pred'].mean()
call_g
```
